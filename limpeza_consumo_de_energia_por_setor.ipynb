```python
def clean_dict(dictd):
    for i in dictd:
        df = dictd[i]
        df.reset_index(inplace=True)
        headers = df.iloc[0].values
        df.columns = headers
        df.drop(index=0, axis=0, inplace=True)
        dictd[i] = df
     
```

```python
import pandas as pd

db = pd.read_excel('databases/Dados EPE/Consumo de Energia por Setor.xlsx', usecols='A:BB').transpose()

dictd_raw = {
    "Setor_Energético": db[0: 17],
    "Setor_Energético_%": db[22: 32]
}

clean_dict(dictd_raw)

def format(string:str, percentage=False):
    words = string.lower().split(' ')
    if percentage:
        words.append("%")
    return "_".join(words)

def format_columns(df, ):
    return df.rename(columns={old_name: format(old_name) for old_name in df.columns})

dictd_raw["Setor_Energético"]  

```
